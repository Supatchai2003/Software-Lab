**Nested Decisions**

```python
x = 5
if x > 2:
  print('Bigger than 2')
  print('Still bigger')
print('Done with 2')

for i in range(5):
  print(i)
  if i > 2:
    print('Bigger than 2')
  print('Done with i',i)
print('All done')
```

3. เขียนโปรแกรม ทายตัวเลข
**กติกา**
1.รับค่าสูงสุด ทางคีย์บอร์ด

2.รับค่าต่ำสุด ทางคีย์บอร์ด

3.ค่าตัวเลขที่เป็นคำตอบ ได้มาจาก built-in function ->random.randint(a,b)

4.การทายตัวเลข สามารถทายตัวเลขได้ 6 ครั้ง

5.ถ้าผู้เล่นทายผิด ให้แสดงว่าค่านั้นมีค่ามากกว่า หรือ น้อยกว่า ตัวเลขที่เป็นคำตอบ

```python
import random
r = 6
high = int(input("ใส่ค่าสูงสุด"))
low = int (input("ใส่ค่าต่ำสุด"))
answer = random.randint(low,high)
while r >=0:
  if r == 0:
    print('จบเกม')
    break
  guess = int(input("ใส่ค่าที่จะเดา"))
  if guess > answer:
    print('ค่าสูงกว่าคำตอบ')
    r -=1
  if guess < answer:
     print('ค่าต่ำกว่าคำตอบ')
     r-=1
  if guess == answer:
    print('คำตอบถูกต้อง')
    r-=7
    break


```
